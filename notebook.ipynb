```
#loading the dataset
import pandas as pd
import numpy as np
data = pd.read_csv('/content/drive/My Drive/hackathon/train.csv')
data.head()
```

```
#storing the names of the images
images = data.iloc[:, 0]
```

```
#converting the series into numpy array
images = images.to_numpy()
```

```
#storing the labels of the images, whether the image is an emergency or not emergency vehicle
labels = data.iloc[:, 1]
```

```
#converting the series into numpy array
labels = labels.to_numpy()
```

```
#converting the images into array
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import array_to_img
inputs = []
path = '/content/drive/My Drive/hackathon/train/'
for i in range(len(images)):
    print(i, 'images loaded')
    image = path + images[i]
    im = load_img(image)
    im = img_to_array(im)
    inputs.append(im)
```

```
#normalising the input values of the array
inputs = np.array(inputs)
inputs = inputs/255.0
```

```
print(inputs.shape)
```

```
#using pre-trained model, MobileNetV2 as the base model(transfer learning)
import tensorflow as tf
base_model = tf.keras.applications.MobileNetV2(input_shape=(224,224,3),
                                               include_top=False,
                                               weights='imagenet')
```

```
#we are not going to train the base model and use it with pre-trained weights
base_model.trainable = False
```

```
base_model.summary()
```

```
#creating the final model
model1 = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])
```

```
model1.summary()
```

```
#compiling the model
model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate = 0.001), loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```
#fitting the training values and training labels
history = model1.fit(inputs, labels, epochs = 10)
```

```
#loading the testing file
df = pd.read_csv('/content/drive/My Drive/hackathon/test.csv')
df.head()
```

```
#storing the names of testing images
testing_images = df.iloc[:,0]
```

```
#converting the series into numpy array
testing_images = np.array(testing_images)
```

```
#converting the images into array
testing_inputs = []
path = '/content/drive/My Drive/hackathon/test/'
for i in range(len(testing_images)):
    print(i, 'images loaded')
    image = path + testing_images[i]
    im = load_img(image)
    im = img_to_array(im)
    testing_inputs.append(im)
```

```
#normalising the values of testing array
testing_inputs = np.array(testing_inputs)
testing_inputs = testing_inputs/255.0
```

```
print(testing_inputs.shape)
```

```
#using the trained model for  prediction on testing inputs
prediction = model1.predict(testing_inputs)
```

```
np.set_printoptions(formatter={'float_kind':'{:f}'.format})
prediction[:5]
```

```
#assigning value 0 when prediction value is less than 0.5 and value 1 when prediction values is greater than 0.5
for i in range(len(prediction)):
    if(prediction[i][0] >= 0.5):
        prediction[i][0] = 1.0
    else:
        prediction[i][0] = 0.0
```

```
prediction = np.array(prediction, dtype = 'int32')
prediction[:10]
```

```
#adding the label column for the training images
df['emergency_or_not'] = prediction
```

```
df.head()
```

```
#writing the dataframe to the csv file
df.to_csv('/content/drive/My Drive/hackathon/test.csv', index = False)
```
